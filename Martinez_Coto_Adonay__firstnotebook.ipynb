```python
import numpy as np
```

```python
A = np.random.rand(10**4, 10**4) # generate a 10^4x10^4 matrix filled with random numbers.
print(f'A shape: {A.shape}, A type: {type(A)}')
```

```python
x = np.random.rand(10**4, 1) # generate a 10^4x1 vector filled with random numbers
print(f'x shape: {x.shape}, x type: {type(x)}')
```

```python
b = A @ x #we use @ for matrices dot product
b.shape
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.scatter(x, b, s=5, alpha=0.5, color='gold')
ax.set_xlabel('x')
ax.set_ylabel('b')
plt.show()
```
