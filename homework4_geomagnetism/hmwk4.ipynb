### Problem 1 - Part A
Calculate the electric field flux through the surface of a 1 m-radius sphere. The electric field is generated by a uniformly charged $1 \mathrm{~cm}^3$ cube with an electric charge density of $1 \mathrm{~C} / \mathrm{m}^3$. Note that the charged cube is at 0.125 m off the center of the sphere.

**SOLUTION:**

The electric flux ($\Phi_E$) through a closed surface is given by Gaussâ€™s Law:

$$
\Phi_E = \oint_S \vec{E} \cdot d\vec{A} = \frac{Q_{enc}}{\epsilon_0}
$$

Where $\Phi_E$ is electric flux (total field passing through the closed surface) and is given in $\mathrm{N} \cdot \mathrm{m}^2 / \mathrm{C}$, $\vec{E}$ is the electric field, $d\vec{A}$ is the differential area vector, $Q_{enc}$ is the total charge enclosed within the surface, and $\epsilon_0$ is the permittivity of free space that is equal to $8.85 \times 10^{-12} \mathrm{~F} / \mathrm{m}$.

The charge density ($\rho$) of the cube is given as $1 \mathrm{~C} / \mathrm{m}^3$, and the volume of the cube is $1 \mathrm{~cm}^3 = 1 \times 10^{-6} \mathrm{~m}^3$. 

The total charge ($Q_{enc}$) in the cube is:

$$
Q_{enc} = \rho \cdot V = (1 \mathrm{~C} / \mathrm{m}^3) \cdot (1 \times 10^{-6} \mathrm{~m}^3) = 1 \times 10^{-6} \mathrm{~C}
$$

For the Gauss's Low the flux depends on the charge enclosed, not on the distance from the charge to the surface. Therefore, even if the cube is not centered in the sphere, the flux through the sphere is:

```python
rho = 1.0  # Charge density (C/m^3)
V = 1.0e-6  # Volume of the cube (m^3)
Q_enc = rho * V  # Total charge enclosed in the cube (C)
epsilon_0 = 8.85e-12  # Permittivity of free space (F/m)

# Compute electric flux
Phi_E = Q_enc / epsilon_0
Phi_E
```

### Problem 1 - Part C
Show that the magnetic field flux through the surface of a sphere is zero if the magnetic field is generated by a dipole. You must directly carry out the surface integration of the magnetic field over the sphere.

**SOLUTION:**


The vector potential for a magnetic dipole $\mathbf{m}$ is given by the equation:

$$
A = \frac{\mu_0}{4\pi} \frac{\mathbf{m} \times \mathbf{\hat{r}}}{r^2}
$$

Where $\mu_0$ is the permeability of free space, $\mathbf{r}$ is the position vector, and $r$ is the distance from the dipole. Also in cartesian coordinates, $\mathbf{\hat{r}}$ is given by:

$$
\mathbf{\hat{r}} = sin(\theta)cos(\phi) \hat{i} + sin(\theta)sin(\phi) \hat{j} + cos(\theta) \hat{k}
$$

Then for \mathbf{m \hat{k}} we have that \mathbf{m} = m(0,0,1), this means that the product $\mathbf{m} \times \mathbf{\hat{r}}$ is given by:

$$
\mathbf{m} \times \mathbf{\hat{r}} = \begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
0 & 0 & m \\
sin(\theta)cos(\phi) & sin(\theta)sin(\phi) & cos(\theta)
\end{vmatrix}
$$

Then we have to calculate the determinant:

$$
\begin{aligned}
\mathbf{m} \times \mathbf{\hat{r}} &= \hat{i} \left(0 - m sin(\theta)cos(\phi)\right) - \hat{j} \left(0 - m sin(\theta)sin(\phi)\right) + \hat{k} \left(0 - 0\right) \\
&= -m sin(\theta)cos(\phi) \hat{i} + m sin(\theta)sin(\phi) \hat{j}\\
&= m \left(sin(\theta)sin(\phi) \hat{j} - sin(\theta)cos(\phi) \hat{i}\right)
\end{aligned}
$$

Now we can write the vector potential as:

$$
\mathbf{A} = \frac{\mu_0}{4\pi} \frac{m \left(sin(\theta)sin(\phi) \hat{j} - sin(\theta)cos(\phi) \hat{i}\right)}{r^2}
$$


For get the magnetic field $\mathbf{B}$, we need to take the curl of the vector potential:

$$
\mathbf{B} = \nabla \times \mathbf{A}
$$

Using spherical coordinates, we have:

$$
\begin{aligned}
\left ( \nabla \times \mathbf{A} \right )_r &= \frac{1}{r sin(\theta)} \left( \frac{\partial}{\partial \theta} \left( A_{\phi} sin(\theta) \right) - \frac{\partial A_{\theta}}{\partial \phi} \right) \\
\left ( \nabla \times \mathbf{A} \right )_{\theta} &= \frac{1}{r} \left( \frac{\partial A_{r}}{\partial \phi} - \frac{\partial}{\partial r} \left( r A_{\phi} \right) \right) \\
\left ( \nabla \times \mathbf{A} \right )_{\phi} &= \frac{1}{r} \left( \frac{\partial}{\partial r} \left( r A_{\theta} \right) - \frac{\partial A_{r}}{\partial \theta} \right)
\end{aligned}
$$

Then separating the components of the vector potential we have:

$$
\begin{aligned}
A_x &= \frac{\mu_0}{4\pi} \frac{-m sin(\theta)cos(\phi)}{r^2} \\
A_y &= \frac{\mu_0}{4\pi} \frac{m sin(\theta)sin(\phi)}{r^2} \\
A_z &= 0
\end{aligned}
$$

Next, for converting to spherical coordinates we have:

$$
\begin{aligned}
\hat{i} &= sin(\theta)cos(\phi) \hat{r} + cos(\theta)cos(\phi) \hat{\theta} - sin(\phi) \hat{\phi} \\
\hat{j} &= sin(\theta)sin(\phi) \hat{r} + cos(\theta)sin(\phi) \hat{\theta} + cos(\phi) \hat{\phi} \\
\hat{k} &= cos(\theta) \hat{r} - sin(\theta) \hat{\theta}
\end{aligned}
$$

Then we can write the vector potential in spherical coordinates:    

$$
\begin{aligned}
A_x &= \frac{\mu_0}{4\pi} \frac{-m sin(\theta)cos(\phi)}{r^2} = \frac{\mu_0}{4\pi} \frac{-m}{r^2} \left( sin(\theta)cos(\phi) \hat{r} + cos(\theta)cos(\phi) \hat{\theta} - sin(\phi) \hat{\phi} \right) \\
A_y &= \frac{\mu_0}{4\pi} \frac{m sin(\theta)sin(\phi)}{r^2} = \frac{\mu_0}{4\pi} \frac{m}{r^2} \left( sin(\theta)sin(\phi) \hat{r} + cos(\theta)sin(\phi) \hat{\theta} + cos(\phi) \hat{\phi} \right) \\
A_z &= 0
\end{aligned}
$$

Now we will expand each term of the curl of the vector potential:

$$
\begin{aligned}
A_r &= - \frac{\mu_0}{4\pi} \frac{-m}{r^2} sin(\theta)sin(\phi) (sin(\theta)cos(\phi) ) + \frac{\mu_0}{4\pi} \frac{m}{r^2} sin(\theta)sin(\phi) (sin(\theta)sin(\phi) ) \\
A_r &= 0 \\
\end{aligned}
$$

$$
\begin{aligned}
A_{\theta} &= - \frac{\mu_0}{4\pi} \frac{-m}{r^2} sin(\theta)cos(\phi) (cos(\theta)cos(\phi) ) + \frac{\mu_0}{4\pi} \frac{m}{r^2} sin(\theta)sin(\phi) (cos(\theta)sin(\phi) ) \\
A_{\theta} &= 0 \\
\end{aligned}
$$

$$
\begin{aligned}
A_{\phi} &= - \frac{\mu_0}{4\pi} \frac{-m}{r^2} sin(\theta)cos(\phi) (-sin(\phi) ) + \frac{\mu_0}{4\pi} \frac{m}{r^2} sin(\theta)sin(\phi) (cos(\phi) ) \\
A_{\phi} &= \frac{\mu_0}{4\pi} \frac{m}{r^2} sin(\theta) sin^2(\phi) + \frac{\mu_0}{4\pi} \frac{m}{r^2} sin(\theta) cos^2(\phi) \\
A_{\phi} &= \frac{\mu_0}{4\pi} \frac{m}{r^2} sin(\theta) \left( sin^2(\phi) + cos^2(\phi) \right) \\
A_{\phi} &= \frac{\mu_0}{4\pi} \frac{m}{r^2} sin(\theta) \\

\end{aligned}
$$

Then we can calculate $B_r$ and $B_{\theta}$, because $B_{\phi}$ is equal to zero:

$$
\begin{aligned}
B_r &= \frac{1}{r sin(\theta)} \left( \frac{\partial}{\partial \theta} \left( A_{\phi} sin(\theta) \right) - \frac{\partial A_{\theta}}{\partial \phi} \right) \\
B_r &= \frac{1}{r sin(\theta)} \left( \frac{\partial}{\partial \theta} \left( \frac{\mu_0}{4\pi} \frac{m}{r^2} sin^2(\theta) \right) - 0 \right) \\
B_r &= \frac{1}{r sin(\theta)} \left( \frac{\mu_0}{4\pi} \frac{m}{r^2} \frac{\partial}{\partial \theta} \left( sin^2(\theta) \right) \right) \\
B_r &= \frac{1}{r sin(\theta)} \left( \frac{\mu_0}{4\pi} \frac{m}{r^2} 2 sin(\theta) cos(\theta) \right) \\
B_r &= \frac{\mu_0}{4\pi} \frac{m}{r^2} \frac{2 cos(\theta)}{r} \\
B_r &= \frac{\mu_0}{4\pi} \frac{2m cos(\theta)}{r^3} \\
\end{aligned}
$$

$$
\begin{aligned}
B_{\theta} &= \frac{1}{r} \left( \frac{\partial A_{r}}{\partial \phi} - \frac{\partial}{\partial r} \left( r A_{\phi} \right) \right) \\
B_{\theta} &= \frac{1}{r} \left( 0 - \frac{\partial}{\partial r} \left( \frac{\mu_0 m sin(\theta)}{4\pi r^2} \right) \right) \\
B_{\theta} &= - \frac{1}{\pi r} \left( \frac{\mu_0 m sin(\theta)}{4} \frac{\partial}{\partial r} \left( \frac{1}{r^2} \right) \right) \\
B_{\theta} &= - \frac{1}{\pi r} \left( \frac{\mu_0 m sin(\theta)}{4} \left( -2 \frac{1}{r^3} \right) \right) \\
B_{\theta} &= \frac{\mu_0 m sin(\theta)}{2 \pi r^4} \\
\end{aligned}
$$

Now we calculate the flux through the sphere:

$$
\Phi_B = \oint_S \vec{B} \cdot d\vec{A} = \oint_S \vec{B} \cdot \hat{n} dA
$$

Where $\hat{n}$ is the unit normal vector to the surface of the sphere. The area element in spherical coordinates is given by:

$$
dA = r^2 sin(\theta) d\theta d\phi
$$

Then we can write the flux as:

$$
\Phi_B = \oint_S \vec{B} \cdot \hat{n} dA = \oint_S \left( B_r \hat{r} + B_{\theta} \hat{\theta} + B_{\phi} \hat{\phi} \right) \cdot \hat{n} dA
$$

Substituting the values of $B_r$ because just the radial component is contributing to the flux:

$$
\Phi_B = \oint_S B_r \hat{r} \cdot \hat{n} dA
$$

Then we can write the flux as:

$$
\Phi_B = \oint_S \frac{\mu_0}{4\pi} \frac{2m cos(\theta)}{r^3} \hat{r} \cdot \hat{n} dA
$$

Where $\hat{n} = \hat{r}$, then we have:

$$
\Phi_B = \oint_S \frac{\mu_0}{4\pi} \frac{2m cos(\theta)}{r^3} dA
$$

Now we can calculate the area of the sphere:

$$
\Phi_B = \frac{\mu_0}{4\pi} \frac{2m}{r^3} \oint_S cos(\theta) dA
$$

The area of the sphere is given by:

$$
A = \oint_S dA = \int_0^{2\pi} \int_0^{\pi} r^2 sin(\theta) d\theta d\phi
$$

Then we can write the flux as:

$$
\Phi_B = \frac{\mu_0}{4\pi} \frac{2m}{r^3} \int_0^{2\pi} \int_0^{\pi} cos(\theta) r^2 sin(\theta) d\theta d\phi
$$

Now we can calculate the integral:

$$
\begin{aligned}
\Phi_B &= \frac{\mu_0}{4\pi} \frac{2m}{r^3} \int_0^{2\pi} d\phi \int_0^{\pi} cos(\theta) sin(\theta) d\theta \\
\Phi_B &= \frac{\mu_0}{4\pi} \frac{2m}{r^3} \int_0^{2\pi} d\phi \left( \frac{1}{2} sin^2(\theta) \right) \Bigg|_0^{\pi} \\
\Phi_B &= \frac{\mu_0}{4\pi} \frac{2m}{r^3} \int_0^{2\pi} d\phi \left( \frac{1}{2} \left( sin^2(\pi) - sin^2(0) \right) \right) \\
\Phi_B &= \frac{\mu_0}{4\pi} \frac{2m}{r^3} \int_0^{2\pi} d\phi \left( \frac{1}{2} \left( 0 - 0 \right) \right) \\
\Phi_B &= 0
\end{aligned}
$$

Then we have that the flux through the sphere is zero.

### Problem 2 - Part A
Calculate the velocity $\mathrm{a}+1$ ion of ${ }^{87} \mathrm{Sr}$ when it is accelerated from zero velocity by an electric potential difference of 10 kV . Assume that the gradient of the electric potential is uni-directional and constant.



**SOLUTION:**

For the principle of the conservation of energy, the kinetic energy ($KE$) gained by the ion is equal to the work done by the electric field on the ion:

$$
KE = q \cdot V
$$

Where $KE$ is kinetic energy in Joules, $q$ is the charge of the ion in Coulombs, and $V$ is the electric potential difference in Volts.

The kinetic energy can also be expressed in terms of the mass ($m$) and velocity ($v$) of the ion $KE = \frac{1}{2}mv^2$.

Setting the two expressions for kinetic energy equal gives:

$$
\frac{1}{2}mv^2 = q \cdot V
$$

Solving for the velocity ($v$):

$$
\begin{align*}
\frac{1}{2}mv^2 &= q \cdot V \\
2qV &= mv^2 \\
\frac{2qV}{m} &= v^2 \\
v &= \sqrt{\frac{2qV}{m}}
\end{align*}
$$

The charge of the ${ }^{87} \mathrm{Sr}$ ion is $q = +1e = 1.6 \times 10^{-19} \mathrm{~C}$, and the mass of the ${ }^{87} \mathrm{Sr}$ ion is the multiplication of the atomic mass of the Sr by the atomic mass unit (u). 1 u = $1.67 \times 10^{-27} \mathrm{~kg}$.

```python
q = 1.602e-19  # Charge of the Sr ion (C)
Sr_mass_u = 87.62  # Atomic mass of Sr (u)
Sr_mass_kg = Sr_mass_u * 1.67e-27  # Mass of Sr ion (kg)
V = 10000  # Potential difference (V)

# Compute velocity
v = (2 * q * V / Sr_mass_kg) ** 0.5
print(f"Velocity of Sr ion: {v:.2e} m/s")
```

When the above ion enters a vacuum chamber with a magnetic field of 0.1 T, it wil follow a circular path because of the Lorentz force. Based on the equal magnitude of the Lorentz force and the centrifugal force, calculate the radius of the circular path.

**SOLUTION:**

The force acting on a charged particle moving in a magnetic field, also known as the Lorentz force ($F_L$), is given by:

$$
F_L = q \cdot v \cdot B
$$

Where $q$ is the charge of the ion, $v$ is the velocity of the ion, and $B$ is the magnetic field strength. 

The centrifugal force ($F_c$) acting on the ion moving in a circular path is given by:

$$
F_c = \frac{mv^2}{r}
$$

Here $m$ is the mass of the ion, $v$ is the velocity of the ion, and $r$ is the radius of the circular path.

- $F_c$ = centrifugal force
- $m$ = mass of the i
Setting the two forces equal gives:

$$
\begin{align*}
F_L &= F_c \\
q \cdot v \cdot B &= \frac{mv^2}{r} \\
r &= \frac{mv}{q \cdot B}
\end{align*}
$$

From the part (a) we have the velocity of the ion and all the other parameters so we can calculate the radius of the circular path.

```python
# Compute radius of circular motion
B = 0.1  # Magnetic field strength (T)
r = (Sr_mass_kg * v) / (q * B)
print(f"Radius of circular motion: {r:.4f} m")
```

### Problem 3 - Part A
Assuming that the geomagnetic field corresponds to a geocentric axial dipole, calculate the latitude of a site where the field inclination is $45^{\circ}$.

**SOLUTION:**

The equation 5.40 for the inclination ($I$) of the geomagnetic field in Lowrie [2007] is given by:

$$
\tan(I) = 2 \cdot tan(\lambda)
$$

Where $\lambda$ is the latitude. Rearranging gives:

$$
\begin{aligned}
\tan\lambda &= \frac{1}{2} \tan{I}
\lambda &= \tan^{-1}\left(\frac{1}{2} \tan{I}\right)
\end{aligned}
$$

Substituting $I = 45^{\circ}$ gives:

$$
\begin{aligned}
\lambda &= \tan^{-1}\left(\frac{1}{2} \tan{45^{\circ}}\right) \\
&= \tan^{-1}\left(\frac{1}{2} \cdot 1\right) \\
&= \tan^{-1}\left(\frac{1}{2}\right) \\
&\approx 26.57^{\circ}
\end{aligned}
$$

### Problem 4 - Part C
What are the local values of inclination and declination of the field?

**SOLUTION:**

From Lowrie [2007] in equations 5.32 the inclination is $$I = \arctan{\left(\frac{B_V}{B_H}\right)}= \arctan{\left(\frac{B_V}{\sqrt{B_N^2 + B_E^2}}\right)}$$ and the declination is given by $$D = \arctan{\left(\frac{B_E}{B_N}\right)}$$.

# Problem 5

Get the 14th generation International Geomagnetic Reference Field data for the 2025-2030 epoch and use the Gauss coefficients for the dipole ( $n=1$ ) and quadrupole ( $) n=2$ ) components of the geomagnetic field for the following questions:

### Problem 5 - Part A
Tabulate the Gauss coefficients for the dipole and quadrupole components of the geomagnetic field from the IGRF model.

```python
igrf_file_path = "igrf14coeffs.txt"

dipole_coeffs = []
quadrupole_coeffs = []

with open(igrf_file_path, "r") as file:
    lines = file.readlines()

# Find the column index for 2025.0
header = lines[3].split()  # The fourth row contains headers
year_index = header.index("2025.0")

# Extract dipole and quadrupole coefficients
for line in lines[4:]:  # Data starts from row 5
    parts = line.split()
    if len(parts) < year_index + 1:
        continue

    g_or_h = parts[0]  # Whether it's g or h
    n = int(parts[1])  # Degree
    m = int(parts[2])  # Order
    value = float(parts[year_index])  # Extract value for 2025

    # Store dipole (n=1) and quadrupole (n=2) coefficients
    if n == 1:
        dipole_coeffs.append((g_or_h, n, m, value))
    elif n == 2:
        quadrupole_coeffs.append((g_or_h, n, m, value))

# Convert extracted data into tables
df_dipole = pd.DataFrame(dipole_coeffs, columns=["Type", "Degree (n)", "Order (m)", "Value (nT)"])
df_quadrupole = pd.DataFrame(quadrupole_coeffs, columns=["Type", "Degree (n)", "Order (m)", "Value (nT)"])


print("Dipole Coefficients for 2025:")
print(df_dipole)

print("\nQuadrupole Coefficients for 2025:")
print(df_quadrupole)

```

### Problem 5 - Part B
Calculate the root mean square intensity of the dipole component of the field at the Earth's surface. The mean squared value, $R_n$, of the intensity of the geomagnetic field at the Earth's surface due to the component of degree $n$ is given by

$$
R_n=(n+1) \sum_{m=0}^n\left(\left(g_n^m\right)^2+\left(h_n^m\right)^2\right)
$$


```python
# Extract Gauss coefficients for the dipole (n=1) from the previously read data
dipole_g = [row[3] for row in dipole_coeffs if row[0] == "g"]  # Extract g_n^m values
dipole_h = [row[3] for row in dipole_coeffs if row[0] == "h"]  # Extract h_n^m values

# Ensure that h_n^0 is considered 0 (since it does not exist for m=0)
dipole_h.insert(0, 0.0)  

# Compute R_1 using the given formula
n = 1  # Dipole degree
R_1 = (n + 1) * sum(g**2 + h**2 for g, h in zip(dipole_g, dipole_h))

print(f"R_1 Calculation Steps:")
print(f"n: {n}")
print(f"g_n^m values: {dipole_g}")
print(f"h_n^m values: {dipole_h}")
print(f"R_1: {R_1:.4f} nT^2")

# Compute the RMS intensity of the dipole component
RMS_dipole_intensity = R_1 ** 0.5
print(f"RMS Dipole Intensity: {RMS_dipole_intensity:.4f} nT")
```

### Problem 5 - Part C
Calculate the corresponding intensity of the quadrupole component and express it as a percentage of the dipole field intensity.

```python
# Extract Gauss coefficients for the quadrupole (n=2) from the previously read data
quadrupole_g = [row[3] for row in quadrupole_coeffs if row[0] == "g"]  # Extract g_n^m values
quadrupole_h = [row[3] for row in quadrupole_coeffs if row[0] == "h"]  # Extract h_n^m values

# Ensure that h_n^0 is considered 0 (since it does not exist for m=0)
quadrupole_h.insert(0, 0.0)

# Compute R_2 using the given formula
n = 2  # Quadrupole degree
R_2 = (n + 1) * sum(g**2 + h**2 for g, h in zip(quadrupole_g, quadrupole_h))

print(f"R_2 Calculation Steps:")
print(f"n: {n}")
print(f"g_n^m values: {quadrupole_g}")
print(f"h_n^m values: {quadrupole_h}")
print(f"R_2: {R_2:.4f} nT^2")

RMS_quadrupole_intensity = R_2 ** 0.5

print(f"RMS Quadrupole Intensity: {RMS_quadrupole_intensity:.4f} nT")

percentage_quadrupole = (RMS_quadrupole_intensity / RMS_dipole_intensity) * 100

print(f"Percentage of Quadrupole Intensity relative to Dipole Intensity: {percentage_quadrupole:.2f}%")
```

\begin{array}{lll}
\hline n & m & P l_n^m(\cos \theta) \\
\hline 0 & 0 & 1 \\
1 & 0 & \cos \theta \\
1 & 1 & \sin \theta \\
2 & 0 & \frac{1}{4}(3 \cos 2 \theta+1)=\frac{3}{2} \cos ^2 \theta-\frac{1}{2} \\
2 & 1 & \frac{\sqrt{3}}{2} \sin 2 \theta=\sqrt{3} \sin \theta \cos \theta \\
2 & 2 & \frac{\sqrt{3}}{2} \sin ^2 \theta \\
\hline\\
&&Table~1: ~Schimidt~ polynomials
\end{array}
\end{align*}


```python
R_Earth = 6.371e6  

# Define latitudes and longitudes for the grid
lat_vals = np.linspace(-90, 90, 100)
lon_vals = np.linspace(-180, 180, 200)
lon_grid, lat_grid = np.meshgrid(lon_vals, lat_vals)

# Convert latitudes to colatitude (theta in radians)
theta_grid = np.radians(90 - lat_grid)
phi_grid = np.radians(lon_grid)

V_geomagnetic_table = np.zeros_like(theta_grid)

# Compute the spherical harmonic terms for n=1 and n=2 and get the geomagnetic potential
for n, g_n, h_n in [(1, [dipole_coeffs[0][3], dipole_coeffs[1][3]], [0, dipole_coeffs[2][3]]),#]:
                     (2, [quadrupole_coeffs[0][3], quadrupole_coeffs[1][3], quadrupole_coeffs[3][3]], 
                         [0, quadrupole_coeffs[2][3], quadrupole_coeffs[4][3]])]:

    for m in range(n + 1):
        if n == 1 and m == 0:
            P_nm = np.cos(theta_grid)
        elif n == 1 and m == 1:
            P_nm = np.sin(theta_grid)
        elif n == 2 and m == 0:
            P_nm = (3/2) * np.cos(theta_grid)**2 - 1/2
        elif n == 2 and m == 1:
            P_nm = np.sqrt(3) * np.sin(theta_grid) * np.cos(theta_grid)
        elif n == 2 and m == 2:
            P_nm = (np.sqrt(3)/2) * np.sin(theta_grid)**2
        else:
            P_nm = 0 
            
        r = R_Earth 
        # Compute geomagnetic potential
        V_geomagnetic_table += R_Earth * ((R_Earth / r) ** (n + 1)) * (g_n[m] * np.cos(m * phi_grid) + h_n[m] * np.sin(m * phi_grid)) * P_nm

# Plot the modified geomagnetic potential using table values
fig, ax = plt.subplots(figsize=(12, 6))
map = Basemap(projection='robin', lon_0=0, resolution='c', ax=ax)
#map = Basemap(projection='cyl', llcrnrlat=-90, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180, resolution='c', ax=ax)

map.drawcoastlines()
map.drawparallels(np.arange(-90., 91., 30.), labels=[1, 0, 0, 0])
map.drawmeridians(np.arange(-180., 181., 60.), labels=[0, 0, 0, 1])

# Convert lat/lon to map coordinates
x, y = map(lon_grid, lat_grid)

# Plot filled contour map
cs = map.contourf(x, y, V_geomagnetic_table, cmap='jet', levels=40)
contours = map.contour(x, y, V_geomagnetic_table, colors='black', linewidths=0.5)

# Add a colorbar
cbar = plt.colorbar(cs, orientation='horizontal', pad=0.1)
cbar.set_label('Geomagnetic Potential (nT)')

plt.title("Geomagnetic Potential using Dipole and Quadrupole Terms (2025 Model)")
plt.show()
```

### Problem 5 - Part E
Find the geomagnetic dipole field vector at Memphis using the IGRF 14 model. Make sure to include the whole procedure you employ in your answer.

**SOLUTION:**

The Earth's magnetic field is often represented using a scalar potential function $W$, which is similar to how we describe gravitational or electrostatic fields. This representation is valid under the assumption that there are no free currents in the region of interest (i.e., outside Earth's core), meaning: 
$$ \nabla \cdot B = 0 $$
Since $B$ is irrotational (curl-free in free space), it can be expressed as the gradient of a potential:
$$ B = -\nabla W $$
where $W$ is the magnetic scalar potential and the negative sign indicates that the field points in the direction of decreasing potential. 

As we bring before Lowrie [2007], in equation 5.34 of his book, the geomagnetic potential $W$ can be expressed as:

$$
\begin{aligned}
W=R \sum_{n=1}^{n=\infty} \sum_{m=0}^{m=n}\left(\frac{R}{r}\right)^{n+1} & \left(g_n^m \cos m \phi +h_n^m \sin m \phi\right) P_n^m(\cos \theta)
\end{aligned}
$$

Remembering that the gradient operator in spherical coordinates is given by:

$$
\nabla = \hat{r} \frac{\partial}{\partial r} + \hat{\theta} \frac{1}{r} \frac{\partial}{\partial \theta} + \hat{\phi} \frac{1}{r \sin \theta} \frac{\partial}{\partial \phi}
$$

Then the geomagnetic field vector can be calculated as:

$$
\begin{aligned}
B_r &= -\frac{\partial W}{\partial r} \\
B_\theta &= -\frac{1}{r} \frac{\partial W}{\partial \theta} \\
B_\phi &= -\frac{1}{r \sin \theta} \frac{\partial W}{\partial \phi}
\end{aligned}
$$

Now, sustituting the expression for $W$ into the equations above:

$$
\begin{aligned}
B_r &= -\frac{\partial}{\partial r} \left(R \sum_{n=1}^{n=\infty} \sum_{m=0}^{m=n}\left(\frac{R}{r}\right)^{n+1} \left(g_n^m \cos m \phi +h_n^m \sin m \phi\right) P_n^m(\cos \theta)\right) \\
B_r &= -\sum_{n=1}^{n=\infty} \sum_{m=0}^{m=n} \left(g_n^m \cos m \phi +h_n^m \sin m \phi\right)P_n^m(\cos \theta) \frac{\partial}{\partial r}  \left(R \left(\frac{R}{r}\right)^{n+1} \right) \\
B_r &= -\sum_{n=1}^{n=\infty} \sum_{m=0}^{m=n} \left(g_n^m \cos m \phi +h_n^m \sin m \phi\right)P_n^m(\cos \theta) \left(R \frac{(n+1) R^{n+1}}{r^{n+2}}\right) \\
\end{aligned}
$$

Since $r=R$ at the surface of the Earth and we are interested in the dipole what means that n=1, we can write:

$$
\begin{aligned}
B_r &= - \sum_{n=1}^{n=\infty} \sum_{m=0}^{m=n} \left(g_n^m \cos m \phi +h_n^m \sin m \phi\right)P_n^m(\cos \theta) (n+1) \\
\end{aligned}
$$

Since in the dipole case $n=1$ and $P_1^0 = \cos \theta$ and $P_1^1 = \sin \theta$ we can write:

$$
\begin{aligned}
B_r &= - 2 \left(g_1^0 \cos \theta + g_1^1 \cos \phi \sin \theta + h_1^1 \sin \phi \sin \theta\right) \\
\end{aligned}
$$

Now for the $B_\theta$ component:

$$
\begin{aligned}
B_\theta &= -\frac{1}{r} \frac{\partial}{\partial \theta} \left(R \sum_{n=1}^{n=\infty} \sum_{m=0}^{m=n}\left(\frac{R}{r}\right)^{n+1} \left(g_n^m \cos m \phi +h_n^m \sin m \phi\right) P_n^m(\cos \theta)\right) \\
B_\theta &= -\frac{R}{r} \sum_{n=1}^{n=\infty} \sum_{m=0}^{m=n} \left(\frac{R}{r}\right)^{n+1} \left(g_n^m \cos m \phi +h_n^m \sin m \phi\right) \frac{\partial}{\partial \theta} P_n^m(\cos \theta) \\
\end{aligned}
$$

From table 1 in part (d) we have that $P_1^0 = \cos \theta$ and $P_1^1 = \sin \theta$, then the the derivative of the $P_1^0$ is $-\sin \theta$ and the derivative of the $P_1^1$ is $\cos \theta$, now substituting this into the equation we have:

$$
B_\theta = -\frac{R}{r} \left(\frac{R}{r}\right)^{2} \left(g_1^0 -\sin \theta + g_1^1 \cos \phi \cos \theta + h_1^1 \sin \phi \cos \theta\right) \\
$$

Then substituting $r=R$ we have:

$$
\begin{aligned}
B_\theta &= -\left(g_1^0 -\sin \theta + g_1^1 \cos \phi \cos \theta + h_1^1 \sin \phi \cos \theta\right) \\
B_\theta &= g_1^0\sin \theta - g_1^1 \cos \phi \cos \theta - h_1^1 \sin \phi \cos \theta \\
\end{aligned}   
$$

And finally for the $B_\phi$ component:

$$
\begin{aligned}
B_\phi &= -\frac{1}{r \sin \theta} \frac{\partial}{\partial \phi} \left(R \sum_{n=1}^{n=\infty} \sum_{m=0}^{m=n}\left(\frac{R}{r}\right)^{n+1} \left(g_n^m \cos m \phi +h_n^m \sin m \phi\right) P_n^m(\cos \theta)\right) \\
B_\phi &= -\frac{R}{r \sin \theta} \sum_{n=1}^{n=\infty} \sum_{m=0}^{m=n} \left(\frac{R}{r}\right)^{n+1} P_n^m(\cos \theta) \frac{\partial}{\partial \phi} \left(g_n^m \cos m \phi +h_n^m \sin m \phi\right) \\
\end{aligned}
$$

Analyzing the derivatives we have that $\frac{\partial}{\partial \phi} cos m \phi = -m \sin m \phi$ and $\frac{\partial}{\partial \phi} sin m \phi = m \cos m \phi$, then substituting this into the equation we have:

$$
B_\phi = -\frac{R}{r \sin \theta} \sum_{n=1}^{n=\infty} \sum_{m=0}^{m=n} \left(\frac{R}{r}\right)^{n+1} P_n^m(\cos \theta) \left(-m g_n^m \sin m \phi + m h_n^m \cos m \phi\right) \\
$$

If we substitute $r=R$ we have:

$$
B_\phi = -\frac{1}{\sin \theta} \sum_{n=1}^{n=\infty} \sum_{m=0}^{m=n} P_n^m(\cos \theta) \left(-m g_n^m \sin m \phi + m h_n^m \cos m \phi\right) \\
$$

In the dipole case $n=1$ and $m=0, 1$, but always $m=0$ the expression is zero, then we have just the $n=1$ and $m=1$ and when P is $P_1^1 = \sin \theta$ we have:

$$
\begin{aligned}
B_\phi &= -\frac{1}{\sin \theta} \left(-g_1^1 \sin \phi + h_1^1 \cos \phi\right) \sin \theta \\
B_\phi &= -\left(-g_1^1 \sin \phi + h_1^1 \cos \phi\right) \\
B_\phi &= g_1^1 \sin \phi - h_1^1 \cos \phi \\
\end{aligned}
$$

Finally, we have the three components of the dipole field vector:

$$
\begin{aligned}
B_r &= - 2 \left(g_1^0 \cos \theta + g_1^1 \cos \phi \sin \theta + h_1^1 \sin \phi \sin \theta\right) \\
B_\theta &= g_1^0\sin \theta - g_1^1 \cos \phi \cos \theta - h_1^1 \sin \phi \cos \theta \\
B_\phi &= g_1^1 \sin \phi - h_1^1 \cos \phi \\
\end{aligned}
$$

Now we can calculate the dipole field vector at Memphis using the IGRF 14 model coefficients that we previously fetched.

```python
# Define Memphis coordinates
lat_memphis = 35.117500
lon_memphis = -89.971107

# Convert to colatitude (theta) and longitude (phi) in radians
theta_memphis = np.radians(90 - lat_memphis)
phi_memphis = np.radians(lon_memphis)

# Extract dipole Gauss coefficients
g_10 = dipole_coeffs[0][3]  # g_1^0
g_11 = dipole_coeffs[1][3]  # g_1^1
h_11 = dipole_coeffs[2][3]  # h_1^1

# Compute magnetic field components
B_r = -2 * (g_10 * np.cos(theta_memphis) + g_11 * np.cos(phi_memphis) * np.sin(theta_memphis) + h_11 * np.sin(phi_memphis) * np.sin(theta_memphis))
B_theta = g_10 * np.sin(theta_memphis) - g_11 * np.cos(phi_memphis) * np.cos(theta_memphis) - h_11 * np.sin(phi_memphis) * np.cos(theta_memphis)
B_phi = g_11 * np.sin(phi_memphis) - h_11 * np.cos(phi_memphis)

# Compute total field intensity
B_total = np.sqrt(B_r**2 + B_theta**2 + B_phi**2)


# Display results
print(f"Magnetic Field Components at Memphis:")
print(f"B_r: {B_r:.4f} nT")
print(f"B_theta: {B_theta:.4f} nT")
print(f"B_phi: {B_phi:.4f} nT")

print(f"Total Magnetic Field Intensity at Memphis: {B_total:.4f} nT")
```
